```python
import openai
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import tkinter as tk

# OpenAI API credentials
openai.api_key = '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'
model_engine = "davinci"  # or any other OpenAI language model

# Search configuration
search_url = "https://sfbay.craigslist.org/search/sss?query="
search_location = "&search_distance=50"
search_max_price = "&max_price=1000"
prompt = "Send a message to the seller to ask if I can pick it up tomorrow: "

def search_craigslist(query):
    driver = webdriver.Chrome()
    driver.get(search_url + query + search_location + search_max_price)

    posts = driver.find_elements(By.CSS_SELECTOR, ".result-row")
    result_text.delete("1.0", tk.END)
    for post in posts:
        title = post.find_element(By.CSS_SELECTOR, ".result-title")
        price = post.find_element(By.CSS_SELECTOR, ".result-price")
        link = title.get_attribute("href")
        result_text.insert(tk.END, f"{title.text} ({price.text})\n{link}\n\n")

def ask_question(prompt, chat_history=""):
    response = openai.Completion.create(
        engine=model_engine,
        prompt=prompt,
        temperature=0.7,
        max_tokens=1024,
        n=1,
        stop=None,
        frequency_penalty=0,
        presence_penalty=0
    )
    answer = response.choices[0].text.strip()
    return answer

def on_submit():
    query = entry.get()
    search_craigslist(query)
    chat_history = ""
    while True:
        question = ask_question(prompt, chat_history)
        if "stop" in question.lower():
            break
        result_text.insert(tk.END, f"Q: {question}\n")
        chat_history += question + "\n"
        answer = ask_question(question, chat_history)
        result_text.insert(tk.END, f"A: {answer}\n\n")
        chat_history += answer + "\n"

root = tk.Tk()
root.title("Craigslist Chatbot")

frame = tk.Frame(root)
frame.pack()

label = tk.Label(frame, text="Search on Craigslist:")
label.pack(side=tk.LEFT)

entry = tk.Entry(frame)
entry.pack(side=tk.LEFT)

submit_button = tk.Button(frame, text="Search", command=on_submit)
submit_button.pack(side=tk.LEFT)

result_text = tk.Text(root, wrap=tk.WORD)
result_text.pack()

root.mainloop()
```

```python
import openpyxl

# Load the Excel file
wb = openpyxl.load_workbook('Profit Margin.xlsx')
sheet = wb.active

# Add Profit column
sheet.insert_cols(4)
sheet['D1'] = 'Profit'

for row in sheet.iter_rows(min_row=2, values_only=True):
    revenue = row[2]
    cost = row[3]
    profit = revenue - cost
    sheet.cell(row=row[0], column=4, value=profit)

# Add Profit Margin column
sheet.insert_cols(5)
sheet['E1'] = 'Profit Margin'

for row in sheet.iter_rows(min_row=2, values_only=True):
    revenue = row[2]
    cost = row[3]
    profit = revenue - cost
    profit_margin = profit / revenue
    sheet.cell(row=row[0], column=5, value=profit_margin)

# Highlight in yellow any profit margin below 0.2
for row in sheet.iter_rows(min_row=2, min_col=5, max_col=5):
    for cell in row:
        if cell.value < 0.2:
            cell.fill = openpyxl.styles.PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

# Filter out those rows
sheet.auto_filter.ref = 'A1:E16'
sheet.auto_filter.add_filter_column(4, ['>=0.2'])

# Create a new sheet to show the average profit margin by store type
ws = wb.create_sheet(title='Average Profit Margin by Store Type')

for row in sheet.iter_rows(min_row=1, max_row=1, values_only=True):
    ws.append(row)

store_types = set([row[0] for row in sheet.iter_rows(min_row=2, min_col=1, max_col=1, values_only=True)])

for store_type in store_types:
    rows = [row for row in sheet.iter_rows(min_row=2, min_col=1, max_col=5, values_only=True) if row[0] == store_type]
    profit_margins = [row[4] for row in rows]
    avg_profit_margin = sum(profit_margins) / len(profit_margins)
    ws.append([store_type, '', '', '', avg_profit_margin])

# Show in cell k18 whether the total sales for stores with chicken nuggets as the best selling item is greater than 32000
for row in sheet.iter_rows(min_row=1, max_row=16, values_only=True):
    if row[6] == 'chicken nuggets':
        if sum([row[2] for row in sheet.iter_rows(min_row=row[0], max_row=row[0], min_col=3, max_col=3, values_only=True)]) > 32000:
            sheet['K18'] = 'Yes'
        else:
            sheet['K18'] = 'No'

# Save the Excel file
wb.save('Profit Margin.xlsx')
```
